```python
#Dependency

import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
```

```python
rates = np.loadtxt('BTC-USD.csv', delimiter = ",", skiprows=1)
rates
```

```python
#Extract 2021 data
close_data = rates[:365]
close_data
```

```python
#calculating Mean 
close_mean = np.mean(close_data)
print("Mean:", close_mean)

#calculating Median 
close_median = np.median(close_data)
print("Median:", close_median)

#calculating Minimum
close_min = np.min(close_data)
print("Minimum:", close_min)

#calculating Maximum 
close_max = np.max(close_data)
print("Maximum:", close_max)

#calculating Standard Deviation 
close_sd = np.std(close_data)
print("Standard Deviation:", close_sd)

#Calculating Interquartile Range 
q3, q1 = np.percentile(close_data, [75, 25])
close_IQR = q3 - q1
print("Interquartile Range:", close_IQR)
```

```python
plt.plot(np.arange(len(close_data)), close_data, color = 'red', linestyle = 'dotted')
plt.title("BTC To USD for 2021")
plt.show()
```

```python
plt.boxplot(np.diff(close_data), vert = 0)
plt.title("Distribution of BTC-to-USD daily price increases")
plt.show()
```

A boxplot is a particular kind of chart that is frequently used in explanatory data analysis in descriptive statistics. Box plots use the averages and quartiles of the data to visually depict the distribution of numerical data and skewness. Box plots display a data set's minimum score, first (lower) quartile, median, third (upper), and maximum scores in a five-number summary.

In our instance, we used a boxplot to pinpoint the outliers and the market's trajectory for BTC to USD in 2021. and determined the price difference on a daily basis while monitoring price increases and decreases. We can also make out some of the outliers, which indicates that on some days the variations were really little or extremely large, causing prices to change dramatically.

```python
marker = dict(marker = 'x', markeredgecolor = 'blue')
plt.boxplot(np.diff(close_data), vert = 0, showmeans = True, meanprops = marker)
plt.title("Distribution of BTC-to-USD daily price increases")
plt.show()
```

```python
min_index = np.argmin(close_data)
max_index = np.argmax(close_data)
print("The lowest price was on the day:", min_index)
print("The highest price was on the day:", max_index)
```

```python
sort_data = np.sort(np.diff(close_data))
q1, q3 = np.percentile(sort_data, [25, 75])
IQR = q3 - q1
upper_limit = q3 + (1.5 * IQR)
lower_limit = q1 - (1.5 * IQR)

print('Number of the outliers in the above boxplot is - ', len(sort_data[(sort_data < lower_limit) | (sort_data > upper_limit)]))
```

Outliers are data observations that are outside the box plot due to having significantly higher or lower values based on the formula above. Any data value that is outside of or greater than 1.5 multiplied by the difference between the Upper and Lower Quartiles is classified as an outlier, whether above or below the whiskers/tails.

We have some outliers for BTC to USD for the period 2021; this indicates that during that time, prices 'over-performed' and 'under-performed' in relation to where the majority of the rest data values are. As a result, prices rose dramatically on some days and fell precipitously on others. When prices are dropped, we have more outliers than when prices are raised. We have 11 outliers in our data for the year 2021.
