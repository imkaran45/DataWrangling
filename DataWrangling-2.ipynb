#### Define a function missingdf(df) with input argument df, which could print out all the columns in data frame df and also the missing value rate for each column. For example, with a total of 1000 rows, if column1 has a missing value in 200 rows and its missing value rate will be 0:2 or 20%. The result of the function missingdf(df) will print the new data frame which has two columns: the column_name and the percent_missing.

##### Answer 2

```python
#Defininig the function to find the missing value rate for each column
def missingdf(df):   
    percent_missing = df.isnull().sum() * 100 / len(df)
    missingValueDf = pd.DataFrame({'column_name': df.columns, 'percent_missing': percent_missing})
    print(missingValueDf.to_string(index = False))

#Calling the missingdf function
missingdf(orderData)
```

###### We have defined the missingdf function to calculate the proportion of the missing value for each column. In the function, we have calculated the missing values based on pandas IsNull function and then divided it by the total length of the column and then added the column name and the proportion into another data frame and printed the newly created data frame. This is the best optimal solution that one can do for solving this kind of problem. 

#### Question 1.3
#### Define a function column_list(df) which could return a list which only contains the numerical column names and another list which only contains the categorical column names.
#### Use describe() function from pandas to print out the statistics for all numerical columns.

##### Answer 1.3

```python
#Defininig the function to find the numerical and categorical columns list
def column_list(df):   
    allColumns = df.columns

    #Finding numerical columns name
    numericalColumns = list(df.select_dtypes([np.number]).columns)

    #Finding categorical columns name
    categoricalColumns = list(set(allColumns) - set(numericalColumns))

    return numericalColumns, categoricalColumns

numericalColumnList, categoricalColumnList = column_list(orderData)
print('-------------------------------------------------------------------')
print("Numerical column names are as below:")
print('-------------------------------------------------------------------')
print(numericalColumnList)

print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
print("Categorical column names are as below:")
print('-------------------------------------------------------------------')
print(categoricalColumnList)
```

```python
#Finding statistics of the numerical columns
orderData.describe(include = [np.number])
```

###### Here we defined the column_list function to perform the activities as asked. First, we collected all the columns of the data frame in a variable, then we collected numerical columns based on the data type of the columns in another variable namely ‘numericalColumns’. Finally, for finding categorical columns we just subtracted numerical columns from all columns and then printed both numerical and categorical column lists. Then we calculated statistics of only numerical columns of the data frame. There is no other way to do this problem and the solution implemented is optimal and feasible.

```python
#Checking if there is any null values.
def check_missing(df):
    null_val = df.isnull().sum()
    percent = 100 * df.isnull().sum()/len(df)
    missing_table = pd.concat([null_val, percent], axis = 1)
    col = missing_table.rename(columns = {0 : 'Num', 1 : 'Rate'})
    return col

# Display columns missing values are under 1%.
print("Number of rows: "+ str(len(customerTransactionData)))
cols = check_missing(customerTransactionData)
cols.sort_values(by = "Rate", ascending = False).head(10)
```

##### You will need to transform the data into a format of: the row represent the unique customer, the column will be all product commodity, the value of the column should be the frequency of the purchase on the particular commodity cross entire data. Name this new dataframe as customer_product_freq

```python
customer_product_freq = customerTransactionData.loc[:, ('customer_id', 'commodity')]
print(customer_product_freq.shape)

temp_df = pd.get_dummies(customer_product_freq['commodity'])
print(temp_df.shape)

customer_product_freq.drop(['commodity'], axis = 1, inplace = True)
customer_product_freq = customer_product_freq.join(temp_df)
del temp_df

print(customer_product_freq.shape)
customer_product_freq = customer_product_freq.groupby('customer_id').sum().reset_index()
print(customer_product_freq.shape)
customer_product_freq.head()
```

#### Question 2.2

##### The itemsets(basket) having length more than 1 and minimum support of 5%

```python
transaction_product1 = transaction_product.groupby('basket_id').sum().reset_index()
transaction_product1['basket_id'] = transaction_product1['basket_id'].astype('str')
transaction_product1.set_index('basket_id', inplace = True)

def encode_commodity(commodity):
    if commodity <= 0:
        return 0
    if commodity >=1:
        return 1

transaction_product1 = transaction_product1.applymap(encode_commodity)
print(transaction_product1.shape)
transaction_product1.head()
```

```python
frequent_itemsets = apriori(transaction_product1, min_support = 0.01, use_colnames=True, verbose = 1)
frequent_itemsets_filter= frequent_itemsets
print(frequent_itemsets_filter.shape)
frequent_itemsets_filter['length'] = frequent_itemsets_filter['itemsets'].apply(lambda x: len(x))
frequent_itemsets_filter = frequent_itemsets_filter.query('length >1 and support>0.05')
frequent_itemsets_filter.head()
```

##### The association rules with minimum support of 2% and having lift more than 1.

```python
#Rules
association_rules_df = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1)
association_rules_df.head()
```

```python
print(association_rules_df[ (association_rules_df['lift'] >1) &
                    (association_rules_df['support'] >=0.02)].shape)
association_rules_df[ (association_rules_df['lift'] >1) &
                    (association_rules_df['support'] >=0.02)].head()
```
